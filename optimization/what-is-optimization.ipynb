```python
import matplotlib.pyplot as plt
import numpy as np

# Super simple scenario: y = m*x
input_data = 4
target_value = 20 # We want our model to learn that parameter should be 5 (since 5*4=20)

# Let's start with a random guess for our parameter
parameter = 2
learning_rate = 0.1 # How much we adjust the parameter each time

print(f"Starting Parameter: {parameter:.2f}")

# Lists to store values for plotting
iterations = []
predictions = []
errors = []
parameters = []

# Let's "train" for a few steps (iterations)
for i in range(100):
    prediction = parameter * input_data
    error = target_value - prediction

    # This is a vastly simplified update rule, not true gradient descent,
    # but it illustrates the idea of adjusting the parameter based on error.
    # In real gradient descent, we'd use the derivative of a loss function.
    # parameter_adjustment = error * input_data * learning_rate # A simple heuristic adjustment
    parameter = parameter + (error * input_data *learning_rate) # Nudge the parameter

    # Store values for plotting
    iterations.append(i + 1)
    predictions.append(prediction)
    errors.append(error)
    parameters.append(parameter)

    print(f"Iteration {i+1}: Prediction={prediction:.2f}, Error={error:.2f}, New Parameter={parameter:.2f}")
    if abs(error) < 0.001:
        print("Converged early!")
        break

print(f"\nFinal Parameter: {parameter:.2f}")

# --- Plotting ---
fig, axs = plt.subplots(3, 1, figsize=(10, 12), sharex=True)

# Plot Prediction vs Iteration
axs[0].plot(iterations, predictions, label='Prediction')
axs[0].axhline(y=target_value, color='r', linestyle='--', label='Ground Truth (Target)')
axs[0].set_ylabel('Value')
axs[0].set_title('Prediction vs. Iteration')
axs[0].legend()
axs[0].grid(True)

# Plot Error vs Iteration
axs[1].plot(iterations, errors, label='Error', color='orange')
axs[1].axhline(y=0, color='r', linestyle='--') # Error should ideally go to 0
axs[1].set_ylabel('Error')
axs[1].set_title('Error vs. Iteration')
axs[1].legend()
axs[1].grid(True)

# Plot Parameter vs Iteration
axs[2].plot(iterations, parameters, label='Parameter', color='green')
axs[2].axhline(y=target_value/input_data, color='r', linestyle='--', label='True Parameter (5)') # The target parameter value
axs[2].set_ylabel('Parameter Value')
axs[2].set_title('Parameter vs. Iteration')
axs[2].set_xlabel('Iteration')
axs[2].legend()
axs[2].grid(True)

plt.tight_layout()
plt.show()
```
